# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
import os
print(os.listdir('../data'))
```

```python
df = pd.read_csv('../data/MachineLearningRating_v3.txt', delimiter='|')
```

```python
df.head()
df.info()
df.describe()
```

# Basic Summaries and Visualizations

1. Descriptive Stats


```python
print(df.columns.tolist())
```

```python
df[['TotalPremium', 'TotalClaims', 'CustomValueEstimate']].describe()
```

2. Loss Ratio Calculation


```python
df['LossRatio'] = df['TotalClaims'] / df['TotalPremium']
df['LossRatio'] = df['LossRatio'].replace([np.inf, -np.inf], np.nan)
```

3. Missing value Heatmap

```python
sns.heatmap(df.isnull(), cbar=False)
```

4. Visualize Loss Ratio by Province, Gender, VehicleType

```python
# Province
df.groupby('Province')['LossRatio'].mean().sort_values(ascending=False).plot(kind='bar', title='Avg Loss Ratio by Province')

# Gender
df.groupby('Gender')['LossRatio'].mean().plot(kind='bar', title='Avg Loss Ratio by Gender')

# Vehicle Type
df.groupby('VehicleType')['LossRatio'].mean().plot(kind='bar', title='Avg Loss Ratio by Vehicle Type')
```

## Bivariate & Multivariate Analysis


 TotalPremium vs TotalClaims (Correlation)

```python
# Scatter plot
sns.scatterplot(data=df, x='TotalPremium', y='TotalClaims')
plt.title('TotalPremium vs TotalClaims')
plt.xlabel('Total Premium')
plt.ylabel('Total Claims')
plt.show()

# Correlation coefficient
correlation = df[['TotalPremium', 'TotalClaims']].corr()
print(correlation)
```

Correlation Matrix for Key Financial Columns

```python
cols = ['TotalPremium', 'TotalClaims', 'CustomValueEstimate', 'SumInsured', 'LossRatio']
corr_matrix = df[cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

 Loss Ratio by Zip Code (Top 20)

```python
zip_loss = df.groupby('PostalCode')['LossRatio'].mean().sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 6))
zip_loss.plot(kind='bar', color='coral')
plt.title('Top 20 Zip Codes by Avg Loss Ratio')
plt.ylabel('Avg Loss Ratio')
plt.xticks(rotation=90)
plt.show()
```

Monthly Trends in Total Claims and Premium

```python
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
monthly = df.groupby(df['TransactionMonth'].dt.to_period('M'))[['TotalPremium', 'TotalClaims']].sum()
monthly.index = monthly.index.to_timestamp()

monthly.plot(figsize=(12, 6))
plt.title("Monthly Trends: Total Premium vs Total Claims")
plt.ylabel("Amount")
plt.xlabel("Month")
plt.grid(True)
plt.show()
```

Outlier Detection

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(y=df['TotalPremium'], ax=ax[0])
ax[0].set_title("Total Premium Outliers")

sns.boxplot(y=df['TotalClaims'], ax=ax[1])
ax[1].set_title("Total Claims Outliers")

sns.boxplot(y=df['CustomValueEstimate'], ax=ax[2])
ax[2].set_title("Custom Value Estimate Outliers")

plt.tight_layout()
plt.show()
```
